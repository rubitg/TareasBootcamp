```python
df_Student=pd.read_csv("students_complete.csv")
df_School=pd.read_csv("schools_complete.csv")
```

```python
# Combine the data into a single dataset
# key?? school_name
data_complete=pd.merge(df_Student, df_School, on='school_name', how='inner') 
```

```python
data_complete
```

```python
data_complete.describe(),df_School.describe()
```

```python
#District Summary
    #Syntax: DataFrame.sum(axis=None, skipna=None, level=None, numeric_only=None, min_count=0, **kwargs)
```

```python
#•	Calculate the total number of schools
number_schools=data_complete['School ID'].nunique()
#number_schools GIVE 15
```

```python
#         Calculate the total number of students
number_students=data_complete['Student ID'].nunique()
# number_students gives 39170
```

```python
#       Calculate the total budget 
total_budget=sum(data_complete['budget'].unique())
total_budget
```

```python
#      Calculate the average math score
avg_math = int(data_complete['math_score'].mean())
avg_math
```

```python
#         Calculate the average reading score
avg_reading = int(data_complete['reading_score'].mean())
avg_reading 
```

```python
#         Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_pass_rate=(avg_math+avg_reading)/2
overall_pass_rate
```

```python
#         Calculate the percentage of students with a passing math score (70 or greater)
num_pass_math7 = data_complete.loc[(data_complete['math_score']>=70),'math_score'].count()
per_pass_math7= (num_pass_math7/number_students)*100
per_pass_math7
```

```python
#         Calculate the percentage of students with a passing reading score (70 or greater)
num_pass_read7 = data_complete.loc[(data_complete['reading_score']>=70),'reading_score'].count()
per_pass_read7= (num_pass_read7/number_students)*100
per_pass_read7
```

```python
#       Create a dataframe to hold the above results
#1 crear el tipo diccionario

district_sum = {"Number of Schools":number_schools,"Total Student":number_students,"Total Budget":total_budget,"Average Math Score":avg_math,"Average Reading Score":avg_reading,"% Passing Math":per_pass_math7,
               "% Passing Reading":per_pass_read7,"% Overall Passing Score":overall_pass_rate}

#2 crear el dataframe quelos contenga con todo y su indice
district_sum_DF =pd.DataFrame((district_sum), index=[7])
district_sum_DF 

# faltarian formatos.. esos al final..
```
